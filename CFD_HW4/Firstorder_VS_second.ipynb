```python
from Nozzle_1D import Nozzle
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
# Roe 1
NORMAL_SHOCK = Nozzle("inputs/Medium_inputs.nml")
NORMAL_SHOCK.CFL = .1
NORMAL_SHOCK.upwind_order = 0
NORMAL_SHOCK.damping_scheme = 2
# RUN MEDIUM SIMULATION
p_ROE_order1,_,_,_,_,_,_ = NORMAL_SHOCK.RUN_SIMULATION(verbose = True)
```

```python

# Roe 2
NORMAL_SHOCK = Nozzle("inputs/Medium_inputs.nml")
NORMAL_SHOCK.CFL = .1
NORMAL_SHOCK.upwind_order = 0
NORMAL_SHOCK.local_timestep = True
NORMAL_SHOCK.damping_scheme = 2
NORMAL_SHOCK.converge_at_second_order = 1e-6
# RUN MEDIUM SIMULATION
p_ROE_order2,_,_,_,_,_,_ = NORMAL_SHOCK.RUN_SIMULATION(verbose = True)
```

```python

# Van Leer 1
MEDIUM_ISENTROPIC = Nozzle("inputs/Medium_inputs.nml")
MEDIUM_ISENTROPIC.CFL = .1
MEDIUM_ISENTROPIC.upwind_order = 0
MEDIUM_ISENTROPIC.damping_scheme = 1
# RUN MEDIUM SIMULATION
p_VanLeer_order1,_,_,_,_,_,_ = MEDIUM_ISENTROPIC.RUN_SIMULATION(verbose = True)
```

```python

# Van Leer 2
MEDIUM_ISENTROPIC = Nozzle("inputs/Medium_inputs.nml")
MEDIUM_ISENTROPIC.CFL = .1
MEDIUM_ISENTROPIC.upwind_order = 1
MEDIUM_ISENTROPIC.damping_scheme = 1
# RUN MEDIUM SIMULATION
p_VanLeer_order2,_,_,_,_,_,_ = MEDIUM_ISENTROPIC.RUN_SIMULATION(verbose = True)

```

```python

# Set up the 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Normal Shock Pressure Comparison (NI=100)", fontsize=18)

# Common x-axis values
x = np.linspace(-1, 1, 102)

# Top row: Roe
axs[0, 0].plot(x, p_ROE_order1, color='tab:red')
axs[0, 0].set_title("Roe 1st Order")
axs[0, 0].set_xlabel("x")
axs[0, 0].set_ylabel("Pressure (kPa)")
axs[0, 0].grid(True)

axs[0, 1].plot(x, p_ROE_order2, color='tab:green')
axs[0, 1].set_title("Roe 2nd Order")
axs[0, 1].set_xlabel("x")
axs[0, 1].set_ylabel("Pressure (kPa)")
axs[0, 1].grid(True)

# Bottom row: Van Leer
axs[1, 0].plot(x, p_VanLeer_order1, color='tab:blue')
axs[1, 0].set_title("Van Leer 1st Order")
axs[1, 0].set_xlabel("x")
axs[1, 0].set_ylabel("Pressure (kPa)")
axs[1, 0].grid(True)

axs[1, 1].plot(x, p_VanLeer_order2, color='tab:orange')
axs[1, 1].set_title("Van Leer 2nd Order")
axs[1, 1].set_xlabel("x")
axs[1, 1].set_ylabel("Pressure (kPa)")
axs[1, 1].grid(True)

# Tight layout with room for suptitle
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig("ROE_VS_VAN")
```
