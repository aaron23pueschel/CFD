```python


# Compute L1 error 
pressure_supercoarse_error_L1 = np.linalg.norm(p_compute_Supercoarse[1:-1] - p_exact_Supercoarse, 1) / (SUPERCOARSE.NI - 1)
pressure_coarse_error_L1 = np.linalg.norm(p_compute_Coarse[1:-1] - p_exact_Coarse, 1) / (COARSE.NI - 1)
pressure_medium_error_L1 = np.linalg.norm(p_compute_Medium[1:-1] - p_exact_Medium, 1) / (MEDIUM.NI - 1)
pressure_fine_error_L1 = np.linalg.norm(p_compute_Fine[1:-1] - p_exact_Fine, 1) / (FINE.NI - 1)

# Compute L error 
pressure_supercoarse_error_Linf = np.linalg.norm(p_compute_Supercoarse[1:-1] - p_exact_Supercoarse, np.inf)
pressure_coarse_error_Linf = np.linalg.norm(p_compute_Coarse[1:-1] - p_exact_Coarse, np.inf)
pressure_medium_error_Linf = np.linalg.norm(p_compute_Medium[1:-1] - p_exact_Medium, np.inf)
pressure_fine_error_Linf = np.linalg.norm(p_compute_Fine[1:-1] - p_exact_Fine, np.inf)

# Compute order of convergence for L1
p0_L1 = np.log(pressure_supercoarse_error_L1 / pressure_coarse_error_L1) / np.log(2)
p1_L1 = np.log(pressure_coarse_error_L1 / pressure_medium_error_L1) / np.log(2)
p2_L1 = np.log(pressure_medium_error_L1 / pressure_fine_error_L1) / np.log(2)

# Compute order of convergence for L
p0_Linf = np.log(pressure_supercoarse_error_Linf / pressure_coarse_error_Linf) / np.log(2)
p1_Linf = np.log(pressure_coarse_error_Linf / pressure_medium_error_Linf) / np.log(2)
p2_Linf = np.log(pressure_medium_error_Linf / pressure_fine_error_Linf) / np.log(2)

# Compute L2 error using np.linalg.norm (norm type 2)
pressure_supercoarse_error_L2 = np.linalg.norm(p_compute_Supercoarse[1:-1] - p_exact_Supercoarse, 2) / np.sqrt(SUPERCOARSE.NI - 1)
pressure_coarse_error_L2 = np.linalg.norm(p_compute_Coarse[1:-1] - p_exact_Coarse, 2) / np.sqrt(COARSE.NI - 1)
pressure_medium_error_L2 = np.linalg.norm(p_compute_Medium[1:-1] - p_exact_Medium, 2) / np.sqrt(MEDIUM.NI - 1)
pressure_fine_error_L2 = np.linalg.norm(p_compute_Fine[1:-1] - p_exact_Fine, 2) / np.sqrt(FINE.NI - 1)

# Compute order of convergence for L2
p0_L2 = np.log(pressure_supercoarse_error_L2 / pressure_coarse_error_L2) / np.log(2)
p1_L2 = np.log(pressure_coarse_error_L2 / pressure_medium_error_L2) / np.log(2)
p2_L2 = np.log(pressure_medium_error_L2 / pressure_fine_error_L2) / np.log(2)


```

```python
x_values = [1,2,4,8]
y_values = [10e-2,20e-2,40e-2,80e-2]
y_values2 = [10e-2,40e-2,160e-2,640e-2]
pressure_L1_error = [pressure_fine_error_L1, pressure_medium_error_L1, pressure_coarse_error_L1, pressure_supercoarse_error_L1]
pressure_L2_error = [pressure_fine_error_L2, pressure_medium_error_L2, pressure_coarse_error_L2, pressure_supercoarse_error_L2]
pressure_Linf_error = [pressure_fine_error_Linf, pressure_medium_error_Linf, pressure_coarse_error_Linf, pressure_supercoarse_error_Linf]

plt.figure(figsize=(10, 6))

# Plot the errors for pressure
plt.loglog(x_values, pressure_L1_error, 'o-', label='Pressure L1 Error', color='b')
plt.loglog(x_values, pressure_L2_error, 's-', label='Pressure L2 Error', color='g')
plt.loglog(x_values, pressure_Linf_error, '^-', label='Pressure L Error', color='r')
plt.loglog(x_values,y_values2,"--",label='Second order line')
plt.loglog(x_values,y_values,"--",label='First order line')
plt.xlabel('h', fontsize=12)
plt.ylabel('Error Magnitude', fontsize=12)
plt.title('Pressure Error ', fontsize=14)
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
x_values = [1,2,4]
# Order of convergence for pressure (L1, L2, L)
pressure_L1 = [p2_L1, p1_L1, p0_L1]
pressure_L2 = [p2_L2, p1_L2, p0_L2]
pressure_Linf = [p2_Linf, p1_Linf, p0_Linf]
plt.title("Pressure order of accuracy")
plt.ylabel("Order of accuracy P")
plt.xlabel("h")
plt.ylim([0,3])
plt.plot(x_values, pressure_L1, 'o-', label='Pressure L1', color='b')
plt.plot(x_values, pressure_L2, 's-', label='Pressure L2', color='g')
plt.plot(x_values, pressure_Linf, '^-', label='Pressure L', color='r')
plt.legend(["L1","L2",r"$L_{\infty}$"])
```
