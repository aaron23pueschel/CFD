```python
from Nozzle_1D import Nozzle
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
tolerance = 1e-12
```

```python
# Initialize MEDIUM ISENTROPIC CASE
MEDIUM_ISENTROPIC = Nozzle("inputs/Medium_inputs.nml")

# RUN MEDIUM SIMULATION
p_compute_Medium,u_compute_Medium,rho_compute_Medium,p_exact_Medium,\
    u_exact_Medium,rho_exact_Medium,convergence_history_Medium = MEDIUM_ISENTROPIC.RUN_SIMULATION(verbose = True,convergence_criteria=tolerance)
```

```python
plt.plot(MEDIUM_ISENTROPIC.x[1:],p_compute_Medium[1:-1])
plt.title("1D nozzle without shocks")
plt.xlabel(r"x")
plt.ylabel("Pressure (kPa)")
```
