## COARSE DOMAIN

```python
p_exact_Coarse = np.array(p_exact_Coarse)
pressure_coarse_error = np.linalg.norm(p_compute_Coarse[1:-1] - p_exact_Coarse) / (COARSE.NI - 1)

p_exact_Medium = np.array(p_exact_Medium)
pressure_medium_error = np.linalg.norm(p_compute_Medium[1:-1] - p_exact_Medium) / (MEDIUM.NI - 1)

p_exact_Fine = np.array(p_exact_Fine)
pressure_fine_error = np.linalg.norm(p_compute_Fine[1:-1] - p_exact_Fine) / (FINE.NI - 1)

#p_exact_Superfine = np.array(p_exact_Superfine)
#pressure_superfine_error = np.linalg.norm(p_compute_Superfine[1:-1] - p_exact_Superfine) / (SUPERFINE.NI - 1)

p1 = np.log(pressure_coarse_error / pressure_medium_error) / np.log(2)
p2 = np.log(pressure_medium_error / pressure_fine_error) / np.log(2)
#p3 = np.log(pressure_fine_error / pressure_superfine_error) / np.log(2)
p2
```

```python
# Density order of accuracy

rho_exact_Coarse = np.array(rho_exact_Coarse)
density_coarse_error = np.linalg.norm(rho_compute_Coarse[1:-1] - rho_exact_Coarse) / (COARSE.NI - 1)

rho_exact_Medium = np.array(rho_exact_Medium)
density_medium_error = np.linalg.norm(rho_compute_Medium[1:-1] - rho_exact_Medium) / (MEDIUM.NI - 1)

rho_exact_Fine = np.array(rho_exact_Fine)
density_fine_error = np.linalg.norm(rho_compute_Fine[1:-1] - rho_exact_Fine) / (FINE.NI - 1)

#rho_exact_Superfine = np.array(rho_exact_Superfine)
#density_superfine_error = np.linalg.norm(rho_compute_Superfine[1:-1] - rho_exact_Superfine) / (SUPERFINE.NI - 1)

p1_rho = np.log(density_coarse_error / density_medium_error) / np.log(2)
p2_rho = np.log(density_medium_error / density_fine_error) / np.log(2)
#p3_rho = np.log(density_fine_error / density_superfine_error) / np.log(2)

p2_rho
```
